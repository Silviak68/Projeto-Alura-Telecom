```python
#Importar os dados
url = 'https://raw.githubusercontent.com/ingridcristh/challenge2-data-science/refs/heads/main/TelecomX_Data.json'
import pandas as pd
import requests
response = requests.get(url)
data = response.json()
df = pd.DataFrame(data)
df
```

Compreender quais informações o dataset possui

Dicionário de dados

•	customerID: número de identificação único de cada cliente

•	Churn: se o cliente deixou ou não a empresa

•	gender: gênero (masculino e feminino)

•	SeniorCitizen: informação sobre um cliente ter ou não idade igual ou maior que 65 anos

•	Partner: se o cliente possui ou não um parceiro ou parceira

•	Dependents: se o cliente possui ou não dependentes

•	tenure: meses de contrato do cliente

•	PhoneService: assinatura de serviço telefônico

•	MultipleLines: assisnatura de mais de uma linha de telefone

•	InternetService: assinatura de um provedor internet

•	OnlineSecurity: assinatura adicional de segurança online

•	OnlineBackup: assinatura adicional de backup online

•	DeviceProtection: assinatura adicional de proteção no dispositivo

•	TechSupport: assinatura adicional de suporte técnico, menos tempo de espera

•	StreamingTV: assinatura de TV a cabo

•	StreamingMovies: assinatura de streaming de filmes

•	Contract: tipo de contrato

•	PaperlessBilling: se o cliente prefere receber online a fatura

•	PaymentMethod: forma de pagamento

•	Charges.Monthly: total de todos os serviços do cliente por mês

•	Charges.Total: total gasto pelo cliente


Normalizando o dataframe

```python
df = pd.json_normalize(data)
df
```

```python
#Verificando se há valores únicos

for col in df.columns:
  print(f'(col): {df[col].unique()}')
  if df[col].nunique()<50:
    print(df[col].unique())
      #separa os itens (colunas) com um tracejado
    print ('-'*50)
```

```python
#verificando se há valores duplicados
numerodeduplicados = df.duplicated().sum()
#outra alternativa = print('numero de duplicados', df.duplicated().sum())
print(f'numero de valores duplicados: ', numerodeduplicados)
```

```python
#Verificando valores nulos
print('número de valores nulos: ', df.isnull().sum())
```

```python
#verficando valores vazios ou em branco
df.apply(lambda x: x.astype(str).str.strip () == '').sum()
```

```python
#transformar coluna account.charges.total para float substituindo valores vazios por zero

#df['account.Charges.Total'] = df['account.Charges.Total'].replace(' ', '0').astype(float)
#df.apply(lambda x: x.astype(str).str.strip () == '').sum()

#os valores ausentes são substituidos por zero e não são reconhecidos como valores nulos
```

```python
# 2º método para transformar coluna account.Charges.Total para float eliminando valores vazios
df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors = 'coerce')
df.apply(lambda x: x.astype(str).str.strip () == '').sum()
df.info()
#os valores ausentes ficam registrados como valores nulos
```

```python
# tratar dados da coluna churn
#sendo o principal dado a analisar, decidimos por retirar as entradas com valores vazios, em vez de buscar algum método de preenchimento
df = df[df['Churn'].str.strip() != '']
#verificamos quantas entradas restaram depois de retirar os valores vazios
print('Número de linhas do Dataframe com valores válidos de churn: ', len(df))

```

```python
#verificando a existência de valores nulos depois da transformação das colunas churn e charges.total
print('número de valores nulos:\n', df.isnull().sum())
```

```python
#Tratando valores nulos remanescentes
#Como a coluna Charges.Total é cumulativa, não faz sentido substituir valores pela média.
#Restam, portanto 2 opções: calcular os valores multiplicando os valores de Charges.Monthly pelos da coluna ternure
#ou, como o número de ocorrências é insignificante, retirar essas linhas do dataset

df = df.dropna(subset = ['account.Charges.Total'])
print('número de valores nulos:\n', df.isnull().sum())
```

Padronizando as variáveis categóricas em português

```python
colunas = {                                                                # Dicionário com os novos nomes das colunas.
    'customerID': 'id',
    'Churn': 'evasão',
    'customer.gender': 'gênero',
    'customer.SeniorCitizen': 'clientes_60+',
    'customer.Partner': 'possui_parceiro',
    'customer.Dependents': 'possui_dependentes',
    'customer.tenure': 'tempo_contrato',
    'phone.PhoneService': 'servico_telefone',
    'phone.MultipleLines': 'multiplas_linhas',
    'internet.InternetService': 'tipo_internet',
    'internet.OnlineSecurity': 'seguranca_online',
    'internet.OnlineBackup': 'backup_online',
    'internet.DeviceProtection': 'protecao_dispositivo',
    'internet.TechSupport': 'suporte_tecnico',
    'internet.StreamingTV': 'streaming_tv',
    'internet.StreamingMovies': 'streaming_filmes',
    'account.Contract': 'tipo_contrato',
    'account.PaperlessBilling': 'fatura_digital',
    'account.PaymentMethod': 'metodo_pagamento',
    'account.Charges.Monthly': 'valor_mensal',
    'account.Charges.Total': 'total_cobrado'
}
df.rename(columns=colunas, inplace=True)

#traduzindo os termos usados nas colunas

df['evasão'] = df['evasão'].replace({'No': 'Não', 'Yes': 'Sim'})
df['gênero'] = df['gênero'].replace({'Female': 'Feminino', 'Male': 'Masculino'})
df['possui_parceiro'] = df['possui_parceiro'].replace({'Yes': 'Sim', 'No': 'Não'})
df['possui_dependentes'] = df['possui_dependentes'].replace({'Yes': 'Sim', 'No': 'Não'})
df['servico_telefone'] = df['servico_telefone'].replace({'Yes': 'Sim', 'No': 'Não'})
df['multiplas_linhas'] = df['multiplas_linhas'].replace({'No': 'Não', 'Yes': 'Sim', 'No phone service': 'Sem serviço de telefone'})
df['tipo_internet'] = df['tipo_internet'].replace({'No': 'Não'})
df.head()
```

Verificando a proporção de clientes evadidos

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
churn_counts = df['evasão'].value_counts()
churn_percentages = (churn_counts / len(df)) * 100

ax = churn_counts.plot(kind='bar')
plt.title('Proporção do Cancelamento de Contratos')
plt.xlabel('Cancelamento ')
plt.ylabel('Quantidade de Clientes')
plt.xticks(rotation=0)

for i, percentage in enumerate(churn_percentages):
    ax.text(i, churn_counts.iloc[i] + 50, f'{percentage:.1f}%', ha='center', va='bottom')

plt.show()
```

Observamos que de cada quatro clientes, mais de um deles encerrou o contrato. A proporção de 26,6% é preocupante, pois demonstra uma grande dificuldade em fidelizar a clientela.

Para ficar mais inteligível, vamos apresentar gráficos para cada categoria em separado


```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10,6))
ax=sns.countplot(x='gênero', hue = 'evasão', data=df)
plt.title('Contagem e Porcentagem de Evasão por Gênero')
plt.xlabel('Gênero')
plt.ylabel('Quantidade de Clientes')
plt.legend(title='Cancelamentos', labels=['Não', 'Sim'], loc='center left', bbox_to_anchor=(1, 0.5))


total_by_genero = df['gênero'].value_counts()


x_tick_labels = [label.get_text() for label in ax.get_xticklabels()]

for p in ax.patches:
    height = p.get_height()

    bar_center_x = p.get_x() + p.get_width() / 2


    category_index = int(round(bar_center_x))
    current_genero = x_tick_labels[category_index]


    total_for_current_genero = total_by_genero[current_genero]


    percentage = (height / total_for_current_genero) * 100

    # mostrar total absoluto e porcentagem
    ax.text(bar_center_x,
            height + 20,
            f'{height:.0f} ({percentage:.1f}%)',
            ha='center',
            va='bottom',
            size=9)

plt.show()
```

Não verificamos diferença significativa na evasão de clientes levando em conta o gênero e, portanto, poodemos descartar essa variável para nossa análise uma vez que os valores são praticamente idênticos.

```python
plt.figure(figsize=(10,6))
ax=sns.countplot(x='clientes_60+', hue = 'evasão', data=df)
plt.title('Contagem e Porcentagem de Evasão de clientes 60+')
plt.xlabel('clientes 60-/ clientes 60+')
plt.ylabel('Quantidade de Clientes')
plt.legend(title='Cancelamentos', labels=['Não', 'Sim'], loc='center left', bbox_to_anchor=(1, 0.5))

total_by_clientes_60 = df['clientes_60+'].value_counts()


x_tick_labels = [label.get_text() for label in ax.get_xticklabels()]

for p in ax.patches:
    height = p.get_height()

    bar_center_x = p.get_x() + p.get_width() / 2



    category_index = int(round(bar_center_x))
    current_category_value = int(x_tick_labels[category_index])


    total_for_current_category = total_by_clientes_60[current_category_value]


    percentage = (height / total_for_current_category) * 100


    ax.text(bar_center_x,
            height + 20,
            f'{height:.0f} ({percentage:.1f}%)',
            ha='center',
            va='bottom',
            size=9)

plt.show()
```

Observamos, nesse gráfico, que a porcentagem de evasão de clientes 60+ é significativamente maior do que a de clientes que têm menos de 60 anos. Enquanto há uma evasão de 41,7% entre os clientes idosos, os clientes com idade menor que 60 anos tem uma taxa de cancelamento de apenas 23,7%. Isso pode ter vários motivos, como o fato de que o ônus do contrato pode ser elevado para uma pessoa aposentada ou a dificuldade da geração mais velha de lidar com novas tecnologias. Porém, considerando que a proporção de idosos entre os clientes é baixa, de apenas cerca de 16%, a evasão de clientes 60+ não teria um peso muito baixo no total de cancelamentos de contrato. Isso significa que esse não é o fator principal a explicar essa evasão.
